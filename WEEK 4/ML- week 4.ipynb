```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

data = pd.read_csv("FuelConsumption.csv")

selected_features = ['ENGINESIZE', 'CYLINDERS', 'FUELCONSUMPTION_COMB', 'CO2EMISSIONS']
data = data[selected_features]

plt.figure(figsize=(12, 8))
plt.scatter(data['ENGINESIZE'], data['CO2EMISSIONS'], color='red', label='Data Points')
plt.title('CO2 Emissions vs Engine Size', color='white', fontsize=16)
plt.xlabel('Engine Size', color='white', fontsize=14)
plt.ylabel('CO2 Emissions', color='white', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7, color='gray')
plt.gca().set_facecolor('black')
plt.tick_params(colors='white', labelsize=12)
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=12)
plt.show()

X = data[['ENGINESIZE']]
y = data['CO2EMISSIONS']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

def evaluate_polynomial_regression(degree):
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    y_test_pred = model.predict(X_test_poly)

    mae = mean_absolute_error(y_test, y_test_pred)
    rss = np.sum((y_test - y_test_pred) ** 2)
    r2 = r2_score(y_test, y_test_pred)

    print(f"Degree {degree} Polynomial Regression:")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Residual Sum of Squares (RSS): {rss:.2f}")
    print(f"R2-Score: {r2:.2f}\n")

plt.figure(figsize=(12, 8))
plt.scatter(data['ENGINESIZE'], data['CO2EMISSIONS'], color='red', label='Data Points')
plt.title('CO2 Emissions vs Engine Size', color='white', fontsize=16)
plt.xlabel('Engine Size (Liters)', color='white', fontsize=14)
plt.ylabel('CO2 Emissions (g/km)', color='white', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7, color='gray')
plt.gca().set_facecolor('black')
plt.tick_params(colors='white', labelsize=12)
plt.legend(facecolor='black', edgecolor='white', labelcolor='white', fontsize=12)
plt.show()

def evaluate_polynomial_regression(degree):
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    y_test_pred = model.predict(X_test_poly)

    mae = mean_absolute_error(y_test, y_test_pred)
    rss = np.sum((y_test - y_test_pred) ** 2)
    r2 = r2_score(y_test, y_test_pred)

    print(f"Degree {degree} Polynomial Regression:")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"Residual Sum of Squares (RSS): {rss:.2f}")
    print(f"R2-Score: {r2:.2f}\n")

    plt.figure(figsize=(12, 8))
    plt.scatter(X, y, color='red', label='Data Points')
    X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    plt.plot(X_range, model.predict(poly.transform(X_range)), color='green', label=f'Degree {degree} Fit')
    plt.title(f'Polynomial Regression (Degree {degree})', color='black', fontsize=16)
    plt.xlabel('Engine Size (Liters)', color='black', fontsize=14)
    plt.ylabel('CO2 Emissions (g/km)', color='black', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7, color='white')
    plt.gca().set_facecolor('black')
    plt.tick_params(colors='black', labelsize=12)
    plt.legend(facecolor='black', edgecolor='yellow', labelcolor='white', fontsize=18)
    plt.show()

for degree in [3, 4, 5]:
    evaluate_polynomial_regression(degree)
```
